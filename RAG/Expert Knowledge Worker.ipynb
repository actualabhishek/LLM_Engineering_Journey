# AITechFarm Knowledge Assistant

## Expert Knowledge Worker

### A question-answering assistant that is an expert knowledge worker
### Built for AITechFarm, an applied-AI technology company
### Uses simple, brute-force RAG (Retrieval Augmented Generation) to answer questions about employees and products

```python
import os
import glob
from pathlib import Path
import gradio as gr
from openai import OpenAI
from dotenv import load_dotenv
```

```python
load_dotenv(override=True)
```

```python
class AitechfarmAssistant:
    def __init__(self, knowledge_folder, model= "gpt-4.1-nano"):
        load_dotenv(override=True)
        self.openai_api_key = os.getenv("OPENAI_API_KEY")
        self.openai = OpenAI(api_key=self.openai_api_key)
        self.model = model
        self.knowledge = {}
        self.system_prefix = """
        You represent AITechFarm, the applied-AI technology company. You are an expert in answering questions about AITechFarm;
        its employees and its products. You are provided with additional context that might be relevant to the user's question. Give brief, accurate answers. If you don't know the answer, say so.
        Relevant context :
        """
        self.load_knowledge(knowledge_folder)

    def load_knowledge(self, folders):
        for folder in folders:
            filenames = glob.glob(folder + "/*")
            for filename in filenames:
                name = Path(filename).stem
                with open(filename, "r", encoding="utf-8") as f:
                    self.knowledge[name.lower()] = f.read()
    def get_relevant_context(self, message):
        text = ''.join(ch for ch in message if ch.isalpha() or ch.isspace())
        text = text.lower()
        results = []
        for name, content in self.knowledge.items():
            if name in text:
                results.append(content)
        return results

    def additional_context(self, message):
        relevant_context = self.get_relevant_context(message)
        if not relevant_context:
            return "There is no additional context relevant to the user's question."
        result = "The following additional context might be relevant in answering the user's question:\n\n"
        result += "\n\n".join(relevant_context)
        return result
    def chat(self, message, history):
        system_message = self.system_prefix + self.additional_context(message)
        messages = [{"role": "system", "content": system_message}] + history + [{"role": "user", "content": message}]
        response = self.openai.chat.completions.create(model=self.model, messages=messages)
        return response.choices[0].message.content
```

```python
assistant = AitechfarmAssistant(["knowledge-base/employees", "knowledge-base/products"])
```
