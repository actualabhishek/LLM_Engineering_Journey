## Overview
This notebook is a hands-on exploration of two core Hugging Face `transformers` concepts: **tokenizers** (how raw text becomes model input) and **pipelines** (the high-level API for running inference on common NLP tasks). It builds intuition by comparing behavior across multiple open-source models rather than just running one in isolation.

## Part 1: Tokenizers

- **Basic tokenization** — encoded a test sentence with the Llama 3.1 tokenizer, then decoded each token ID individually to see how BPE splits words into subword pieces (including leading-space markers).
- **Vocabulary size** — checked `len(tokenizer.vocab)` to see the full token vocabulary Llama 3.1 was trained with (128,256 tokens).
- **Chat templates** — used `apply_chat_template()` to see how a `system` + `user` message list gets converted into the exact prompt format a chat-tuned model expects, including role markers and end-of-turn tokens.
- **Cross-model tokenizer comparison** — encoded the *same sentence* with four different tokenizers side by side:
  | Model | Tokenizer style |
  |---|---|
  | **Llama 3.1 (8B)** | `<\|start_header_id\|>` / `<\|eot_id\|>` role markers |
  | **Phi-4-mini** | `<\|System\|>` / `<\|end\|>` role markers |
  | **DeepSeek-V3.1** | `<｜User｜>` / `<｜Assistant｜>` markers, includes a `</think>` tag |
  | **Qwen2.5-Coder (7B)** | Tested specifically on a Python code snippet, to see how a *code-specialized* tokenizer splits syntax (indentation, parentheses, function names) differently from natural language |

  This comparison highlights that every model family defines its own special tokens and chat formatting — there's no universal prompt format across LLMs.

## Key Takeaways
- **Tokenizer ≠ universal**: each model family has its own vocabulary, subword splitting behavior, and special tokens — code shouldn't assume portability between tokenizers.
- **Chat templates encode model-specific conventions**: the same `messages` list produces structurally different prompts per model, which is why `apply_chat_template()` exists rather than hand-writing prompts.
- **Pipelines abstract away boilerplate**: for standard tasks (sentiment, NER, summarization, translation, classification, generation), `pipeline(task, model=...)` handles tokenization, inference, and decoding in one call — versus the more manual tokenize → generate → decode flow needed for custom/seq2seq setups like the T5 QA example.
- **Library version sensitivity**: pipeline task names and class names (e.g. `AutoModelWithLMHead` → `AutoModelForSeq2SeqLM`) have changed across `transformers` versions — worth pinning versions or checking release notes when tutorials/code don't match installed behavior.
