# 🏢 Company Brochure Generator

**An AI-powered tool that turns a company name and URL into a polished, professional brochure — automatically.**

---

### 🔍 How it works
This notebook uses the **Anthropic Claude API** with **built-in web search** to research a company in real time, then streams a fully formatted **Markdown brochure** directly into the notebook as it's generated.

### ✨ Key Features
- **Zero manual research** — just provide a company name and URL
- **Live web search** grounds the brochure in current, accurate information
- **Streaming output** — watch the brochure being written in real time
- **Clean Markdown formatting** — ready to copy, export, or share

### 🛠️ Tech Stack
`Python` · `Anthropic Claude API` · `Web Search Tool` · `Jupyter Notebook`

---

```python
import os
import anthropic
from dotenv import load_dotenv
from IPython.display import display, Markdown, update_display
from proto.marshal.compat import message
```

```python
load_dotenv(override=True)
```

```python
anthropic_api_key = os.getenv("ANTHROPIC_API_KEY")
if anthropic_api_key and anthropic_api_key.startswith("sk-ant") and len(anthropic_api_key) >10:
    print("anthropic api key detected")
else:
    print("anthropic api key not detected")
```

```python
MODEL = "claude-sonnet-4-6"
```

```python
anthropic_client = anthropic.Anthropic(api_key=anthropic_api_key)
```

```python
SYSTEM_PROMPT = '''
You are an expert business analyst and copywriter.
When given a company name and URL, you will:
1. Use web search to explore the company's website, about page, careers, culture, and any other relevant pages.
2. Synthesize everything into a short, compelling company brochure in Markdown.
3. The brochure should cover: what the company does, its culture/values, who its customers are, and career opportunities if available.
4. Write in clean Markdown without code blocks.
'''
```

```python
def streaming_brochure_generator(company_name, url):
    print(f'Researching {company_name}....Claude will search the web automatically \n')
    with anthropic_client.messages.stream(
        model=MODEL,
        max_tokens=2500,
        system=SYSTEM_PROMPT,

        tools=[{"type": "web_search_20260209", "name": "web_search", "max_uses": 5}],
        messages=[{"role": "user", "content": f"Research {company_name} at {url} and create a professional brochure"}]
    ) as stream:
        full_response = ""
        display_handler = display(Markdown(""), display_id=True)
        for text in stream.text_stream:
            full_response += text
            update_display(Markdown(full_response), display_id=display_handler.display_id)
```

```python
if __name__ == '__main__':
    user_input = input("Enter company name and URL separated by a comma\n  example: Hugging Face, https://huggingface.co :")
    print(f'\n{user_input}\n')
    parts = user_input.split(",", 1)
    if len(parts) != 2:
        print("Invalid input. Please enter a company name and URL separated by a comma")
    else:
        company_name = parts[0].strip()
        url = parts[1].strip()
        streaming_brochure_generator(company_name, url)
```
